```python
import torch

from fmdiffae.data.data_utils import get_webdataset
import glob
import webdataset as wds
import os
def get_webdataset(
    split="train",
    base_dir="/data/hai-res/shared/datasets/mtg-jamendo/processed",
    data_type="spec",
    shuffle_size=2048,
):
    shard_paths = sorted(glob.glob(os.path.join(base_dir, split, f"{data_type}-*.tar")))

    if split == "train":
        dataset = wds.WebDataset(
            shard_paths, resampled=True
        ).shuffle(shuffle_size)
    else:
        dataset = wds.WebDataset(shard_paths, resampled=False)

    dataset = (
        dataset.decode()
        .to_tuple(f"__key__")
        .map(lambda x: x[0])
    )

    return dataset


ds = get_webdataset()

dataloader = torch.utils.data.DataLoader(ds, batch_size=512, num_workers=4)
```
